# Example 4: Explore patent data

This notebook provides insights on how to use the tool to fetch patent data from PubChem.

# Import modules

```python
import plotly

from pyBiodatafuse import id_mapper
from pyBiodatafuse.analyzer.explorer.patent import get_patent_from_pubchem, plot_patent_summary

plotly.io.renderers.default = "notebook_connected"
```

```python
chemicals_of_interest = """glucose
Aspirin"""
chem_list = chemicals_of_interest.split("\n")
```

```python
pubchem_df, pubchem_metadata = id_mapper.pubchem_xref(identifiers=chem_list, identifier_type="name")
```

```python
patent_df = get_patent_from_pubchem(pubchem_df)
```

```python
patent_df
```

# Select specific entry to plot

Since these plotting functions are at server expensive, we recomment to only looks at 1 entity summary at time.

```python
plot_patent_summary(patent_df, compound_id="CID:5793")
```

```python
plot_patent_summary(patent_df, compound_id="CID:2244", interactive=False)
```
